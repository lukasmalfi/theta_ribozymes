```python
import subprocess
import pandas as pd
import numpy as np
```

```python


run=["first","second","third","fourth"]

#run rnarobo for each motif (each iteration and each base: C,T,G,A)
bases=["C","T","G","A"]
for nr in run:
    for bas in bases: 
        inp="/mnt/mnemo6/lukas/anaconda3/bin/rnarobo-2.1.0-linux64 -c --nratio 0.1 /mnt/mnemo6/lukas/Python/ribozymes/genome/combined_v3/falsepositive/"+nr+"/"+nr+"_"+bas+".txt /mnt/mnemo6/lukas/Python/ribozymes/genome/combined_v3/combined.fna >> /mnt/mnemo6/lukas/Python/ribozymes/genome/combined_v3/falsepositive/"+nr+"/"+bas+"/comb_rnarobo_result_precise.txt"
        subprocess.run(inp, shell=True)
```

```python
#Identify and cut out all the viral sequences with a ribozyme from the combined FASTA file (first version)


# Loop through the bases in the "bases" list
for bas in bases: 
    # Create an empty string to store the complete FASTA file
    complete_fasta=""
    
    # Define the file paths for the ribozyme results and the combined FASTA file
    filepath="/mnt/mnemo6/lukas/Python/ribozymes/genome/combined_v3/falsepositive/first/"+bas+"/comb_rnarobo_result_precise.txt"
    filepath_fasta="/mnt/mnemo6/lukas/Python/ribozymes/genome/combined_v3/combined.fna"
    
    # Open the ribozyme results file
    f=open(filepath)
    f_l=f.readlines()
    
    # Initialize variables to store the sequence names and sequences
    seqnam=[]
    sequence=[]
    counter=0
    
    # Loop through each line of the ribozyme results file
    for line in f_l[11:-6]:
        # If the line does not start with "|", it contains the sequence name
        if not line[0]=="|":
            line=line.strip().split()
            seqnam.append(line[2])
            counter=1
        # If the line starts with "|", it contains the sequence itself
        elif counter==1:
            line=line.strip()
            sequence.append(line)
            counter=0

    # Get the unique sequence names
    unique_seqnam=set(seqnam)
    
    # Loop through the unique sequence names
    for seqname in unique_seqnam:
        # Define the input command for sed to extract the sequence from the combined FASTA file
        inp= "sed -n '/"+seqname+"/,/>/p' "+filepath_fasta
        # Run the input command using the subprocess.run() function
        l=subprocess.run(inp,shell=True,capture_output=True,text=True)
        # Add the output to the complete FASTA file string
        complete_fasta+=l.stdout
    
    # Split the complete FASTA file string into individual sequences
    com_fas=complete_fasta.split("\n")
    fasta_corr=[]
    checker=0
    fasta_part=""
    
    # Loop through each line of the complete FASTA file
    for line in com_fas[:-1]:
        # If the line is empty, skip it
        if line=="": 
            continue
        # If the line starts with ">", it contains the sequence name
        elif line[0]==">":
            # If the checker is 0, add the previous sequence to the list of corrected FASTA sequences
            if checker==0:
                fasta_corr.append(fasta_part)
                fasta_part=""
                line=line.split()
                fasta_corr.append(line[0])
                checker=1
            # If the checker is 1, set it to 0
            else:
                checker=0
        # If the line does not start with ">", it contains the sequence itself
        else:
            fasta_part+=line.strip()

    # Add the last sequence to the list of corrected FASTA sequences
    fasta_corr.append(fasta_part)
    
    # Join the corrected FASTA sequences into a single string
    fast_seq="\n".join(fasta_corr)

    # Replace any periods in the corrected FASTA sequences with underscores
    fast_seq=fast_seq.replace(".","_")

    # Write the corrected FASTA sequences to a file
    f= open("/mnt/mnemo6/lukas/Python/ribozymes/genome/combined_v3/falsepositive/first/"+bas+"/combined_all_viral_sequences_v2.fasta","w+")
    f.write(fast_seq)
    f.close()    
```
